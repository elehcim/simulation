```python
bbox = [[-6000.0, 6000.0], [-6000.0, 6000.0], [-6000.0, 6000.0]]
unit_base = {
    'length': (1.0, 'kpc'),
    'velocity': (1.0, 'km/s'),
    'mass': (10**10, 'Msun')
}
```

```python
ds = yt.load("snapshot_0065", unit_base=unit_base, bounding_box=bbox)
```

```python
ds.field_list
```

```python
print ("Redshift =", ds.current_redshift)
```

```python
p = yt.ProjectionPlot(ds, "y", "density")
```

```python
p.zoom(5.0)
p.show()
```

```python
s = yt.SlicePlot(ds, "z", ["density", "velocity_magnitude"], center="max")
s.set_cmap("velocity_magnitude", "kamae")
s.zoom(100.0)
```

```python
s.annotate_velocity()
```

```python
s.annotate_contour("temperature")
```

```python
sc = yt.create_scene(ds)
```

```python
sc.camera.set_width(ds.quan(20, 'kpc'))
```
