```python
import os
# sim = simulation.Simulation(os.path.dirname(snap)) #SIMPATH)
sim = simulation.Simulation(SIMPATH)
```

```python
import ipywidgets

snap_slider = ipywidgets.IntSlider(min=0,max=len(sim)-1,step=1,value=0, continuous_update=False, description='Snap:')

def k(i):
    snap = sim[i]
    snap.properties['boxsize'] = pynbody.units.Unit("1000 Gpc")
    print(len(snap.s))
    color_plot(snap, width=10, title='{:.2f}'.format(snap.header.time), center=True, mag_filter=32, vmin=30, vmax=33)

w = ipywidgets.interactive(k, i=snap_slider)
w
```
