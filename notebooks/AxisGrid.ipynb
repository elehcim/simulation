```python
import pynbody
import numpy as np
import matplotlib.pyplot as plt
from luminosity import surface_brightness
from mpl_toolkits.axes_grid1 import AxesGrid
```

```python
snap = "/home/michele/sim/MoRIA/M1-10_Verbeke2017/M10sim41001/snapshot_0036"
s = pynbody.load(snap)
```

```python
bands_available = ['u', 'b', 'v', 'r', 'i', 'j', 'h', 'k']
nbands = len(bands_available)
```

```python
pynbody.analysis.halo.center(s.star)
# fig, (ax_u, ax_v, ax_b, ax_i) = plt.subplots(ncols=4)
fig, (ax_u, cb_u) = plt.subplots(ncols=2)
sb_u = surface_brightness(s.s, band='u', width=8, resolution=500, subplot=ax_u, cax=cb_u)
# sb_v = surface_brightness(s.s, band='v', width=8, resolution=500, subplot=ax_v)
# sb_b = surface_brightness(s.s, band='b', width=8, resolution=500, subplot=plt, contour=5)
# sb_i = surface_brightness(s.s, band='i', width=8, resolution=500, subplot=plt, contour=5)
```

```python
fig = plt.figure()
grid = AxesGrid(fig, 111,  # similar to subplot(142)
                nrows_ncols=(1, 2),
                axes_pad=0.7,
#                 share_x=True,
#                 share_all=False,
                label_mode="all",
#                 cbar_location="bottom",
                cbar_mode="each",
                cbar_location="top",
#                 cbar_mode="each",
                cbar_size="3%",
                cbar_pad="2%"
               )
sb=list()
for i in range(2):
    sb.append(surface_brightness(s.s, band=bands_available[i], width=8, resolution=500, contour=5,
                                 mag_filter=None, subplot=grid[i], cax=grid.cbar_axes[i]))
```

```python
pynbody.analysis.halo.center(s.star)
fig = plt.figure(figsize=(12,8), dpi=200)
grid = AxesGrid(fig, 111, nrows_ncols=(2, 4),
                axes_pad=0.5,
                label_mode="L",
                cbar_mode="each",
                cbar_location="top",
                cbar_size="3%",
                cbar_pad="2%"
               )
sb=list()
for i in range(nbands):
    sb.append(surface_brightness(s.s, band=bands_available[i], width=8, resolution=500,
                                 isophotes=(22, 26), label_contour=False, gaussian_sigma=0.1,
                                 mag_filter=None, subplot=grid[i], cax=grid.cbar_axes[i]))
```

```python
pynbody.analysis.halo.center(s.star)
fig = plt.figure(figsize=(12,8), dpi=200)
grid = AxesGrid(fig, 111, nrows_ncols=(2, 4),
                axes_pad=0.2,
                label_mode="L",
                cbar_mode="single",
                cbar_location="right",
                cbar_size="3%",
                cbar_pad="2%"
               )
sb=list()
for i in range(nbands):
    sb.append(surface_brightness(s.s, band=bands_available[i], width=8, resolution=500,
                                 isophotes=(22, 26), label_contour=False, gaussian_sigma=0.1,
                                 mag_filter=None, subplot=grid[i], cax=grid.cbar_axes[i], vmin=18, vmax=28))
```
