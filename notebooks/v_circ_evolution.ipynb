```python
import simulation
import pynbody
import matplotlib.pylab as plt
import numpy as np
%matplotlib inline
```

```python
sim = simulation.MoriaSim('69002_p200.0_a600.0_r600.0_c8.15_z0', kicked=True)
```

```python
load_from_file=True
fig, ax_v_circ = plt.subplots(1, figsize=(6,4), dpi=200)
eps = 0.03
for i in range(0, len(sim), 5):
    sim[i]['eps'] = pynbody.array.SimArray(eps*np.ones_like(sim[i]['mass']), pynbody.units.kpc)
    pynbody.analysis.halo.center(sim[i])
    snap = sim[i].g
    p = pynbody.analysis.profile.Profile(sim[i], load_from_file=load_from_file, nbins=100, max=50)
    y = 'v_circ'
    snap_time_gyr = sim[i].properties['time'].in_units("Gyr")
    ax_v_circ.plot(p['rbins'], p[y], color=plt.cm.viridis(i/len(sim)), label='{:.2f} Gyr'.format(snap_time_gyr))
    ax_v_circ.set_xlabel("r ({})".format(p['rbins'].units))

    try:
        ax_v_circ.set_ylabel("$v_c$ [{}]".format(r"${}$".format(getattr(p[y],'units','').latex())))
    except AttributeError:
        ax_v_circ.set_ylabel("{} [{}]".format(y, getattr(p[y],'units','')))
    p.write()
    #     ax_v_circ.legend()
    #     title = '{}   ($t={:5.2f}$ Gyr, snap={})'.format(y, snap_time_gyr, i)
#             ax.set_title(title)
#     ax_v_circ.plot()
```
