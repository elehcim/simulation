To produce something like [here](http://pynbody.github.io/pynbody/tutorials/pictures.html#multi-band-images-of-stars)

```python
import pynbody
import matplotlib.pyplot as plt
```

load the snapshot (which you can download from [here](https://www.dropbox.com/s/t6urvybvh2094xr/snapshot_0400?dl=0))

```python
s = pynbody.load('/home/michele/sim/Birmingham_dataset/snapshot_filaments/snapshot_0400')
```

```python
s.properties
```

Set an high value as boxsize so that errors in rendering are not thrown

```python
s.properties['boxsize'] = pynbody.units.Unit("1 Gpc")
```

Center and rotate

```python
pynbody.analysis.angmom.faceon(s)
```

```python
pynbody.plot.stars.render(s, width=20)
plt.show()
```
